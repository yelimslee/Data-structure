# **`동적 계획법`이란?**

**`동적 계획법`**은 복잡한 문제를 간단한 하위 문제들로 나누어 풀어나가는 알고리즘 설계 방법입니다. 이는 큰 문제를 효과적으로 해결하기 위해 작은 하위 문제들의 해결책을 활용하며, 이로 인해 중복되는 계산을 최소화하고 문제 해결의 속도를 향상시키는 효과를 가지고 있습니다. **`동적 계획법`**은 두 가지 핵심 특성, 즉 ‘최적 부분 구조’와 ‘중복되는 하위 문제’를 기반으로 합니다.

이를 좀 더 이해하기 쉽게 설명드리면, **주어진 문제를 여러 개의 작은 하위 문제로 나누어 풀되, 이러한 하위 문제들의 해결책을 결합해 최종적인 문제를 해결하는 방식**입니다. 그리고 우리는 각 하위 문제의 해결책을 **저장해두고, 나중에 같은 하위 문제를 만나게 되면 이미 저장된 해결책을 활용해 불필요한 계산을 줄입니다.** 즉, 이는 **같은 문제를 단 한 번만 풀도록 하는 알고리즘**이라고 할 수 있습니다.

**`동적 계획법`**은 아래의 두 가지 조건을 만족할 때 적용할 수 있습니다.

- **중복되는 하위 문제(Overlapping Sub-problems)** : 큰 문제를 작은 문제로 분해할 수 있고, 이 작은 문제들이 중복해서 발견된다.
- **최적 부분 구조(Optimal Substructure)** : 작은 문제에서 도출한 해결책이 그것을 포함하는 큰 문제에서도 그대로 유효합니다. 즉, 작은 문제의 해결책을 큰 문제에도 그대로 활용할 수 있다.

---

## **`동적 계획법`의 적용 사례**

**`동적 계획법`**은 굉장히 다양한 문제에 적용할 수 있는 기법입니다. 특히, 최단 경로 문제, 최장 공통 부분 수열 문제, 배낭 문제, 행렬 곱셈 순서 결정 문제 등에서 그 유용성이 입증되었습니다. 더불어, 동적 계획법은 그래프 알고리즘, 문자열 처리, 순열과 조합, 게임 이론 등 다양한 분야에서 활용되고 있습니다.

---

## **`동적 계획법`의 장점**

- 중복되는 계산을 피함으로써 효율적인 계산을 실현하고, 문제 해결의 속도를 향상시킵니다.
- 최적 부분 구조를 이용해 문제의 최적해를 도출합니다.
- 작은 하위 문제의 해결 결과를 저장하고 재활용함으로써 메모리를 효율적으로 사용합니다.

---

## **`동적 계획법`의 단점**

- 문제의 복잡성과 어려움: **`동적 계획법`**은 복잡한 문제를 해결하기 위해 작은 하위 문제로 분할하는 전략을 채택합니다. 그러나 이러한 분할과 하위 문제를 정의하는 과정은 어렵고, 최적 부분 구조를 파악하는 것도 힘들 수 있습니다. 따라서 **`동적 계획법`**을 적용하기 위해서는 문제에 대한 깊은 이해와 분석 능력이 필요합니다.
- 모든 문제에 적용할 수 없음: **`동적 계획법`**은 모든 종류의 문제에 적용할 수 있는 것은 아닙니다. 일부 문제는 최적 부분 구조나 중복되는 하위 문제가 없을 수 있어, 이런 경우 **`동적 계획법`**의 적용이 불가능하거나 비효율적일 수 있습니다. 이럴 때는 다른 알고리즘 설계 기법을 고려해야 합니다.

# **`메모이제이션`이란?**

메모이제이션은 프로그래밍에서 사용되는 중요한 기법 중 하나입니다. 이 기법은 이전에 계산한 결과를 기억하고 저장하여, 동일한 계산을 다시 수행하지 않도록 함으로써 실행 시간을 단축시킵니다. 그래서 프로그램이 동일한 입력값에 대해 함수를 여러 번 호출할 때, 메모이제이션을 통해 함수의 반환 값을 빠르게 재사용할 수 있습니다. 결과적으로 중복 계산을 제거하여 알고리즘의 실행 시간을 획기적으로 개선할 수 있습니다.

메모이제이션은 주로 재귀 호출과 같이 동일한 함수가 여러 번 호출되는 경우에 활용되며, 큰 문제를 작은 하위 문제로 나누는 동적 계획법에서도 사용됩니다.

---

## **`메모이제이션`의 장점**

### **1. 중복 계산을 피할 수 있습니다.**

어떤 알고리즘은 동일한 입력에 대해 여러 번 계산이 수행되는 경우가 많습니다. 예를 들어 피보나치 수열이나 그래프 탐색 알고리즘에서는 동일한 부분 문제를 여러 번 해결해야 하는 경우가 많습니다. 이런 상황에서 메모이제이션을 활용하면 중복된 계산을 피할 수 있습니다. 즉, 이미 계산한 결과를 저장하고 필요할 때 다시 사용함으로써, 불필요한 반복 계산을 줄일 수 있습니다.

### **2. 실행 시간을 줄여 효율성을 향상시킵니다.**

메모이제이션을 사용하면 이전에 계산한 결과를 재사용할 수 있으므로, 새로운 계산을 수행하는 것보다 더 빠르게 결과를 얻을 수 있습니다. 이는 프로그램의 실행 시간을 크게 단축시키며, 이는 사용자 경험을 향상시키는 데 큰 도움이 됩니다. 특히 반복 계산이 많이 필요한 알고리즘에서는 메모이제이션의 효과를 더욱 잘 느낄 수 있습니다.

### **3. 복잡한 문제를 효과적으로 해결할 수 있습니다.**

메모이제이션은 큰 문제를 작은 부분 문제로 분해해 해결하는 동적 계획법과 밀접한 연관이 있습니다. 메모이제이션을 통해 각 부분 문제의 결과를 저장하고 재사용함으로써, 전체 문제를 효율적으로 해결할 수 있습니다. 이를 통해 복잡한 문제도 쉽게 해결할 수 있습니다.

### **4. 코드의 가독성과 유지 보수성을 향상시킵니다.**

메모이제이션은 코드의 가독성과 유지 보수성을 향상시키는 데에도 큰 도움을 줍니다. 이전에 계산한 결과를 저장하고 재사용하는 방식은 코드를 간결하고 이해하기 쉽게 만듭니다. 이로 인해 코드의 복잡성을 줄이고 로직을 명확하게 표현할 수 있습니다. 이는 코드의 유지 보수를 용이하게 하며, 추후 코드 변경이나 업데이트 작업을 훨씬 쉽게 만듭니다.

## **`메모이제이션`의 활용 예시**

메모이제이션의 활용 사례로 피보나치 수열의 계산을 들 수 있습니다. 피보나치 수열은 이전 두 항의 값을 더해서 다음 항을 구하는 방식의 수열입니다. 일반적으로 재귀 함수를 사용해서 피보나치 수열을 계산하면, 같은 값을 여러 번 계산해야하는 문제가 발생합니다. 예를 들어, **`fibonacci(5)`**를 계산하기 위해서는 **`fibonacci(4)`**와 **`fibonacci(3)`**을 계산해야 합니다. 여기서 **`fibonacci(4)`**는 이미 이전에 계산되었던 값입니다. 이처럼 동일한 하위 문제가 여러 번 반복되는 경우, 메모이제이션을 사용하여 중복 계산을 피할 수 있습니다.

아래의 코드에서 **`memo`**를 사용하여 이전에 계산한 값을 저장합니다. 함수 호출 시 **`memo`**에 해당 값이 저장되어 있는지 확인하며, 값이 있으면 저장된 값을 반환하고, 없으면 재귀적으로 함수를 호출하여 값을 계산한 후 **`memo`**에 저장합니다. 이렇게 함으로써 중복 계산을 피하고 계산 속도를 향상시킬 수 있습니다.

```python
def fibonacci(n, memo={}):
    # 이미 계산된 값이 메모에 있는 경우, 해당 값을 반환
    if n in memo:
        return memo[n]

    # 피보나치 수열의 첫 번째와 두 번째 수는 1
    if n <= 2:
        result = 1
    else:
        # n-1번째와 n-2번째 피보나치 수를 재귀적으로 호출하여 합산
        result = fibonacci(n - 1, memo) + fibonacci(n - 2, memo)

    # 계산된 피보나치 수를 메모에 저장
    memo[n] = result
    return result

# 예제 호출
n = 10
fib_result = fibonacci(n)
print(f"{n}번째 피보나치 수열은: {fib_result}")
```

피보나치 수열의 10번째 항을 계산하는 예시를 살펴보겠습니다.

```python
print(fibonacci(10))
```

해당 코드를 실행하면 **`fibonacci(10)`**의 결과로 55가 반환됩니다. 이때, **`fibonacci(10)`**을 계산하기 위해 **`fibonacci(9)`**와 **`fibonacci(8)`**을 계산해야 합니다. 하지만 **`memo`**에 해당 값들이 이미 저장되어 있으므로 중복 계산을 피하고 즉시 결과를 얻을 수 있습니다. 이처럼 메모이제이션을 사용하면 복잡한 문제를 효율적으로 해결하면서 동시에 코드의 가독성을 높일 수 있습니다.

# **`상향식` 접근법이란?**

상향식 DP는 큰 문제를 해결하기 위해 먼저 작은 부분 문제들의 해답을 찾아내는 방법입니다. 이 과정에서 작은 부분 문제의 결과를 저장하고, 이를 이용해 더 큰 문제의 답을 도출해냅니다.

## **`상향식` 접근법의 특징**

- 반복문을 사용하여 순차적으로 문제를 해결합니다:상향식 접근법은 반복문을 사용해 작은 문제부터 큰 문제로 점진적으로 답을 구해나갑니다. 작은 문제의 해를 도출하고 이를 활용하여 큰 문제의 해를 찾아가는 과정을 거칩니다.
- 중복 계산을 방지합니다:작은 부분 문제의 해를 저장하고 이를 활용하면 중복 계산을 피할 수 있습니다. 계산 결과를 배열이나 테이블에 저장하면, 추후 필요한 부분 문제의 해를 즉시 참조할 수 있습니다.
- 알고리즘 진행 과정이 명확합니다:작은 문제부터 시작해 큰 문제로 해결을 이어가기 때문에, 알고리즘의 진행 과정이 직관적이며 명확하게 이해할 수 있습니다.

# **`하향식` 접근법이란?**

**`하향식` 접근법**은 큰 문제를 작은 부분 문제로 분해하여 해결하는 방식입니다. 이 접근법은 재귀 호출과 메모이제이션 기법을 활용하여 구현됩니다. 큰 문제를 해결하기 위해 필요한 작은 부분 문제의 해를 계산하고, 필요시 저장하여 중복 계산을 방지합니다.

## **`하향식` 접근법의 특징**

- 재귀 호출과 메모이제이션을 사용합니다:재귀 호출을 이용해 큰 문제를 작은 부분 문제로 분해하고, 메모이제이션을 통해 이미 계산한 작은 부분 문제의 해를 저장합니다.
- 구조적인 해결을 제공합니다:큰 문제를 작은 부분 문제로 분할해 해결하기 때문에, 문제의 구조와 해결 순서를 명확하게 이해할 수 있습니다.
- 필요한 부분 문제만을 해결합니다:메모이제이션을 이용해 이전에 계산한 작은 부분 문제의 해를 저장하고 재활용함으로써, 필요한 부분 문제만을 해결하게 되어 계산이 효율적입니다.
- 최적 부분 구조를 활용합니다:하향식 접근법은 ‘큰 문제의 최적해를 작은 부분 문제의 최적해로 구성’하는 문제에 적합합니다.

# **예제로 살펴보는 `상향식` 접근법**

## **피보나치 수열 계산**

피보나치 수열은 각 항이 바로 앞의 두 항을 더한 값으로 이루어진 수열입니다. 상향식 접근법을 활용하면, 작은 항부터 시작하여 이전 항의 결과를 이용해 다음 항을 계산하게 됩니다. 이 과정을 반복함으로써 피보나치 수열의 값을 도출해낼 수 있습니다.

아래의 코드는 **`상향식`** 접근법을 사용하여 피보나치 수열을 계산하는 예시입니다.

```python
def fibonacci(n):
    # 피보나치 수열 초기 값 설정
    fib = [0, 1]

    # 작은 값부터 시작하여 이전 항의 계산 결과를 이용하여 다음 항을 계산
    for i in range(2, n + 1):
        fib.append(fib[i - 1] + fib[i - 2])

    # n번째 항의 값을 반환
    return fib[n]

n = 10
result = fibonacci(n)
print(f"{n}번째 피보나치 수열은: {result}")
```

1. 피보나치 수열 계산을 위해 **`fib`** 리스트를 초기화합니다. 초기 값으로 **`[0, 1]`**을 설정합니다.
2. 반복문을 사용하여 작은 값부터 시작하여 **`fib`** 리스트에 이전 항의 계산 결과를 저장합니다. 각 항은 바로 앞의 두 항을 더한 값으로 계산됩니다.
3. 반복문이 종료되면 **`n`**번째 항의 값을 **`fib[n]`**으로 반환합니다.

이 코드는 작은 값부터 시작하여 피보나치 수열의 값을 계산하므로, 상향식 접근법을 따르고 있습니다. 이전 항의 계산 결과를 이용하여 다음 항을 계산하므로 중복 계산을 피하고 효율적으로 결과를 얻을 수 있습니다.

예를 들어, **`fibonacci(6)`**을 호출하면 **`fib`** 리스트는 **`[0, 1, 1, 2, 3, 5, 8]`**이 되며, **`fib[6]`**의 값인 8이 반환됩니다. 이처럼 상향식 접근법을 사용하여 작은 부분 문제의 해를 계산하고 이를 활용하여 전체 문제의 해를 구하는 방식으로 피보나치 수열을 계산할 수 있습니다.

# **예제로 살펴보는 `하향식` 접근법**

## **피보나치 수열 계산**

**`하향식`** 접근법을 사용하면 큰 문제를 작은 부분 문제로 나누어 해결할 수 있습니다. 재귀 호출과 메모이제이션을 활용하여 작은 부분 문제의 해를 계산하고 이를 이용하여 큰 문제의 해를 구합니다.

아래의 코드는 **`하향식`** 접근법을 사용하여 피보나치 수열을 계산하는 예시입니다.

```python
def fibonacci(n, memo={}):
    # 메모이제이션을 활용하여 이미 계산한 값이 있는 경우 그 값을 반환
    if n in memo:
        return memo[n]

    # 피보나치 수열 초기 값 설정
    if n <= 1:
        result = n
    else:
        # 작은 문제로 분할하여 재귀 호출하고 결과를 저장
        result = fibonacci(n - 1, memo) + fibonacci(n - 2, memo)

    # 계산한 값을 메모이제이션에 저장
    memo[n] = result

    # n번째 항의 값을 반환
    return result


n = 10
result = fibonacci(n)
print(f"{n}번째 피보나치 수열은: {result}")
```

이 코드는 주어진 피보나치 수열 계산 함수를 **`하향식`** 접근법으로 개선한 예시입니다. 이 함수는 **`memo`**라는 매개변수를 활용하여 메모이제이션을 구현합니다. 메모이제이션을 통해 이미 계산한 값이 있다면 그 값을 바로 반환하고, 그렇지 않다면 작은 문제로 분할하여 재귀 호출을 실행하고 그 결과를 저장합니다. 계산 결과를 메모이제이션에 저장한 후, n번째 피보나치 수의 값을 반환합니다. 이를 통해 재귀 호출과 메모이제이션을 활용한 하향식 접근법으로 피보나치 수열을 계산할 수 있습니다.

# **최적 부분 구조**

**최적 부분 구조**는 어떤 문제의 최적의 해결책이 그 문제를 구성하는 하위 문제들의 최적의 해결책들로 이루어져 있다는 개념을 나타냅니다. 이를테면, 전체 문제를 해결하기 위한 방법을 찾는 과정에서, 그 문제를 이루는 작은 부분들의 최적의 해결 방안을 찾아낼 수 있어야 합니다. 그리고 이 작은 부분들의 최적 해결책을 결합하면, 전체 문제에 대한 최적의 해결책을 찾아낼 수 있습니다.

이 개념을 적용하는 대표적인 예시는 최단 경로를 찾는 문제입니다. 예를 들어, 여러 도시를 연결하는 그래프에서 A도시에서 D도시로 가는 최단 경로를 찾는다고 해 봅시다. 이 때, 최적 부분 구조의 개념에 따르면, A에서 D로 가는 최단 경로를 찾는 문제는 A에서 B로 가는 최단 경로, B에서 C로 가는 최단 경로, 그리고 C에서 D로 가는 최단 경로를 찾는 여러 하위 문제로 분할할 수 있습니다.

# **중복 부분 문제**

**중복 부분 문제는, 큰 문제를 해결하는 과정에서 동일한 작은 문제들이 여러 차례 재발하는 상황을 말합니다.** 마치 퍼즐을 맞추는 과정에서 같은 조각이 여러 차례 필요한 것처럼, 이는 큰 문제를 해결하기 위해 동일한 작은 문제를 반복적으로 다루어야 한다는 의미입니다.

이 개념을 이해하는 데 가장 좋은 예시는 **`피보나치 수열`**입니다.

> 피보나치 수열이란, 첫 번째와 두 번째 항이 각각 1이며, 그 이후의 모든 항이 바로 앞 두 항의 합으로 이루어진 수열입니다.
>

```python
def fib(n):
    if n <= 2:
        return 1
    return fib(n - 1) + fib(n - 2)
```

### **문제 해결 접근법**

이 문제는 동적 계획법을 활용하여 효율적으로 해결할 수 있습니다.

- 작은 문제로 분할하기: n개의 계단을 오르는 경우의 수를 구하기 위해, 작은 문제로 나눌 수 있습니다. 마지막 계단에서부터 거꾸로 생각해보면, (n-1)번째 계단을 오르는 경우와 (n-2)번째 계단을 오르는 경우로 나눌 수 있습니다.
    - 위 예시를 풀어서 직접 생각해 봅시다. 8개의 계단을 오르는 경우를 계산해야 한다면, 거꾸로 8번째의 계단에 도달하는 경우부터 생각해 봅시다.
    - 8번째의 계단에 올라오는 경우는
        1. 6번째 계단을 올라온 상황에서 2계단을 올라오는 경우
        2. 7번째 계단을 올라온 상황에서 1계단을 올라오는 경우
    - 이 두가지를 합쳤을때와 동일합니다. 이 말은 즉, **`6번째 계단을 오를 수 있는 모든 경우의 수 + 7번째 계단을 오를 수 있는 모든 경우의 수`**입니다.
    - 점화식으로 표현한다면, **problems(n) = problems(n - 2) + problems(n - 1)로 표현됩니다.**
- 중복 부분 문제 해결하기: 작은 문제로 나눴을 때, 같은 작은 문제를 반복해서 풀어야 하는 상황이 발생할 수 있습니다. 이런 중복 부분 문제를 효율적으로 해결하기 위해, 결과를 저장해 놓을 배열이나 메모리 공간을 활용합니다.
    - 작은 문제 정의에서 본 예시처럼, 8번째 계단을 오르는 경우는 6번째(n - 2)의 경우 + 7번째(n - 1)의 경우입니다. 그렇다면 부분 문제인 6번째 계단을 오르는 경우는 어떤 방식으로 해결할 수 있을까요?
    - 결국 6번째 계단을오르는 경우도 4번째(n - 2)의 경우와 5번째(n - 1)의 경우의 합으로 결과를 구할 수 있습니다. 동일한 방식으로 해결할 수 있는 경우를 중복 부분 문제라고 표현합니다.

---

```python
def count_ways(n):
    if n == 1:
        return 1
    if n == 2:
        return 2

    # 계단의 수가 증가할 때마다 가능한 경우의 수를 저장하기 위한 배열
    dp = [0] * (n + 1)
    dp[1] = 1
    dp[2] = 2

    for i in range(3, n + 1):
        # (n-1)번째 계단을 오르는 경우와 (n-2)번째 계단을 오르는 경우의 합을 저장
        dp[i] = dp[i - 1] + dp[i - 2]

    return dp[n]

n = 5
result = count_ways(n)
print(f"{n}개의 계단을 오르는 방법은 {result}가지 입니다.")
```

위의 코드는 **`n=5`**인 경우, 5개의 계단을 오르는 방법의 수를 계산합니다. 상향식 접근법을 사용하여 작은 문제의 결과를 저장하고, 마지막 계단까지 올라가면서 가능한 모든 경우의 수를 계산하여 결과를 반환합니다. 위 코드의 출력은 다음과 같습니다.

```python
"8개의 계단을 오르는 방법은 5가지 입니다."
```

따라서, 5개의 계단을 오르는 방법은 총 8가지입니다.

이와 같이 동적 계획법을 활용하여 계단 오르기 문제를 효율적으로 해결할 수 있습니다. 동적 계획법은 중복되는 계산을 피하고 작은 문제의 결과를 저장하여 큰 문제를 해결하는 방식이므로, 다양한 문제에 적용하여 효율적인 알고리즘을 개발할 수 있습니다.
